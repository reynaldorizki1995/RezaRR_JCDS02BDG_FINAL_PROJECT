```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('clean_listing.csv')
```

```python
pd.set_option('max_columns',None)
df.head()
```

```python
df.drop(['latitude', 'longitude'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df_dummies = pd.get_dummies(df,drop_first=True)
df_dummies.head()
```

## 1. Default Parameter 

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
import lightgbm as lgb
from yellowbrick.regressor import ResidualsPlot


features = df_dummies.drop('price',axis = 1)
target = df_dummies.price
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state = 101)
```

```python
### build evaluation function

def evaluate(model, test_features, test_labels):
    vals = dict()
    predictions = model.predict(test_features)
    MAE = mean_absolute_error(test_labels, predictions)
    RMSE = np.sqrt(mean_squared_error(test_labels, predictions))
    r2 = r2_score(test_labels, predictions)
    print('Model Performance')
    print(f"MAE: {round(MAE, 3)}")
    print(f"RMSE: {round(RMSE, 3)}")
    print(f"r2: {round(r2, 3)}")
    vals['MAE'] = MAE
    vals['RMSE'] = RMSE
    vals['R2'] = r2
    return vals
```

### 1.1 Linear Regression

```python
linear_regression = LinearRegression()

linear_regression.fit(X_train, y_train)
linear_regression_acc = evaluate(linear_regression, X_test, y_test)

visualizer = ResidualsPlot(linear_regression)
visualizer.fit(X_train, y_train)
visualizer.score(X_test, y_test)
visualizer.show()
```

### 1.2 Lasso

```python
lasso = Lasso()

lasso.fit(X_train, y_train)
lasso_acc = evaluate(lasso, X_test, y_test)

visualizer = ResidualsPlot(lasso)
visualizer.fit(X_train, y_train)
visualizer.score(X_test, y_test)
visualizer.show()
```

### 1.3 Ridge

```python
ridge = Ridge()

ridge.fit(X_train, y_train)
ridge_acc = evaluate(ridge, X_test, y_test)

visualizer = ResidualsPlot(ridge)
visualizer.fit(X_train, y_train)
visualizer.score(X_test, y_test)
visualizer.show()
```

### 1.4 Random Forest Regressor

```python
random_forest = RandomForestRegressor()

random_forest.fit(X_train, y_train)
random_forest_acc = evaluate(random_forest, X_test, y_test)

visualizer = ResidualsPlot(random_forest)
visualizer.fit(X_train, y_train)
visualizer.score(X_test, y_test)
visualizer.show()
```

### 1.5 XGBoost Regressor

```python
booster = XGBRegressor()

booster.fit(X_train, y_train)
booster_acc = evaluate(booster, X_test, y_test)

visualizer = ResidualsPlot(booster)
visualizer.fit(X_train, y_train)
visualizer.score(X_test, y_test)
visualizer.show()
```

### 1.6 LightGBM

```python
lgbm = LGBMRegressor()

lgbm.fit(X_train, y_train)
lgbm_acc = evaluate(lgbm, X_test, y_test)

visualizer = ResidualsPlot(lgbm)
visualizer.fit(X_train, y_train)
visualizer.score(X_test, y_test)
visualizer.show()
```
