## 1. Default Parameter 

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
import lightgbm as lgb
from yellowbrick.regressor import ResidualsPlot


features = df_dummies.drop('price',axis = 1)
target = df_dummies.price
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state = 101)
```

```python
### build evaluation function

def evaluate(model, test_features, test_labels):
    vals = dict()
    predictions = model.predict(test_features)
    MAE = mean_absolute_error(test_labels, predictions)
    MSE = mean_squared_error(test_labels, predictions)
    RMSE = np.sqrt(mean_squared_error(test_labels, predictions))
    r2 = r2_score(test_labels, predictions)
    print('Model Performance')
    print(f"MAE: {round(MAE, 3)}")
    print(f"MSE: {round(MSE, 3)}")
    print(f"RMSE: {round(RMSE, 3)}")
    print(f"r2: {round(r2, 2)}")
    vals['MAE'] = MAE
    vals['MSE'] = MSE
    vals['RMSE'] = RMSE
    vals['R2'] = r2
    return vals
```

### 1.1 Linear Regression

```python
linear_regression = LinearRegression()

linear_regression.fit(X_train, y_train)
linear_regression_acc = evaluate(linear_regression, X_test, y_test)

visualizer = ResidualsPlot(linear_regression)
visualizer.fit(X_train, y_train)
visualizer.score(X_test, y_test)
visualizer.show()
```

### 1.2 Lasso

```python
lasso = Lasso()

lasso.fit(X_train, y_train)
lasso_acc = evaluate(lasso, X_test, y_test)

visualizer = ResidualsPlot(lasso)
visualizer.fit(X_train, y_train)
visualizer.score(X_test, y_test)
visualizer.show()
```

### 1.4 Random Forest Regressor

```python
random_forest = RandomForestRegressor()

random_forest.fit(X_train, y_train)
random_forest_acc = evaluate(random_forest, X_test, y_test)

visualizer = ResidualsPlot(random_forest)
visualizer.fit(X_train, y_train)
visualizer.score(X_test, y_test)
visualizer.show()
```

### 3. Experiment by Feature Selection

In this experiment i choose features by looking the correlation with the target, the minimum number of correlation is **0.1**

```python
#Correlation Ratio (eta) is a measurement of correlation between categorical column and continous column
def correlation_ratio(categories, measurements):
    fcat, _ = pd.factorize(categories)
    cat_num = np.max(fcat)+1
    y_avg_array = np.zeros(cat_num)
    n_array = np.zeros(cat_num)
    for i in range(0,cat_num):
        cat_measures = measurements[np.argwhere(fcat == i).flatten()]
        n_array[i] = len(cat_measures)
        y_avg_array[i] = np.average(cat_measures)
    y_total_avg = np.sum(np.multiply(y_avg_array,n_array))/np.sum(n_array)
    numerator = np.sum(np.multiply(n_array,np.power(np.subtract(y_avg_array,y_total_avg),2)))
    denominator = np.sum(np.power(np.subtract(measurements,y_total_avg),2))
    if numerator == 0:
        eta = 0.0
    else:
        eta = np.sqrt(numerator/denominator)
    return eta
```

```python
for col in df.columns:
    if col not in ['id','latitude','longitude','price']:
        x = correlation_ratio(df[col], df['price'])
        if (x == 0.1) or (x > 0.1):
            print(f'Correlation Between {col} and Price is : {x}')
```
